```python
# Import necessary libraries
import random
import geopandas as gpd
from shapely.geometry import Point
import rasterio
from rasterio.windows import Window
import numpy as np
import pandas as pd
import torch
import torchvision.transforms as transforms
from torch.utils.data import DataLoader, TensorDataset
from torch import nn, optim
from tqdm import tqdm
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from torchvision.transforms import ToPILImage, ToTensor
from torchvision import models  # Import models from torchvision
from torchvision.models import ResNet50_Weights, ResNet101_Weights

# Set random seeds for reproducibility
random.seed(42)
np.random.seed(42)
torch.manual_seed(42)

# Load your administrative shapes and MPI data
print("Loading administrative shapes and MPI data...")
admin_shapes = gpd.read_file('shapefiles/nga_admbnda_adm1_osgof_20190417.shp') 
mpi_data = pd.read_csv('nigeria_mpi_only.csv')

# Function to sample unique coordinates within a geometry
def sample_unique_coordinates(geometry, num_samples):
    minx, miny, maxx, maxy = geometry.bounds
    coordinates = set()
    attempts = 0
    max_attempts = num_samples * 10  # To avoid infinite loops

    while len(coordinates) < num_samples and attempts < max_attempts:
        pnt = Point(random.uniform(minx, maxx), random.uniform(miny, maxy))
        if geometry.contains(pnt) and pnt not in coordinates:
            coordinates.add(pnt)
        attempts += 1

    if len(coordinates) < num_samples:
        print(f"Could only sample {len(coordinates)} points for region.")
    return list(coordinates)

# Sample coordinates for each region
print("\nSampling coordinates for each region...")
sampled_coordinates_per_region = {}
for index, row in admin_shapes.iterrows():
    geometry = row['geometry']
    region_name = row['ADM1_EN']  # Adjust column name as per your data
    sampled_coords = sample_unique_coordinates(geometry, 20)
    sampled_coordinates_per_region[region_name] = sampled_coords

print("\nSampled coordinates per region:")
for region, coords in sampled_coordinates_per_region.items():
    print(f"{region}: {len(coords)} points")

regions = list(sampled_coordinates_per_region.keys())

# Split regions into training and test sets
print("\nSplitting regions into training and test sets...")
train_regions, test_regions = train_test_split(regions, test_size=0.2, random_state=42)

print(f"Training regions ({len(train_regions)}): {train_regions}")
print(f"Test regions ({len(test_regions)}): {test_regions}")

# Function to extract images and labels
def extract_images_and_labels(coords_with_labels, image_path, size=224):
    images = []
    labels = []
    with rasterio.open(image_path) as src:
        for coord_label in coords_with_labels:
            row, col = src.index(coord_label['coordinates'][0], coord_label['coordinates'][1])
            window = Window(col - size // 2, row - size // 2, size, size)
            try:
                image = src.read(window=window)
                if image.shape[1] != size or image.shape[2] != size:
                    print(f"Skipping image at {coord_label['coordinates']} due to incorrect size {image.shape}")
                    continue
                images.append(image)
                labels.append(coord_label['region_name'])
            except Exception as e:
                print(f"Error reading image at {coord_label['coordinates']}: {e}")
                continue
    return images, labels

sat_image_path = 'Nigeria_2018/Nigeria_2018_Combined.tif'

# Extract images and labels for training regions
print("\nExtracting images and labels for training regions...")
train_images = []
train_labels = []
for region in train_regions:
    coords = sampled_coordinates_per_region[region]
    region_coords_labels = [{'coordinates': (point.x, point.y), 'region_name': region} for point in coords]
    images_region, labels_region = extract_images_and_labels(region_coords_labels, sat_image_path)
    train_images.extend(images_region)
    train_labels.extend(labels_region)

# Extract images and labels for test regions
print("\nExtracting images and labels for test regions...")
test_images = []
test_labels = []
for region in test_regions:
    coords = sampled_coordinates_per_region[region]
    region_coords_labels = [{'coordinates': (point.x, point.y), 'region_name': region} for point in coords]
    images_region, labels_region = extract_images_and_labels(region_coords_labels, sat_image_path)
    test_images.extend(images_region)
    test_labels.extend(labels_region)

print(f"\nTotal training images: {len(train_images)}")
print(f"Total test images: {len(test_images)}")

# Map regions to MPI scores
print("\nMapping regions to MPI scores...")
# Apply square root transformation to MPI scores
mpi_data['transformed_mpi'] = np.sqrt(mpi_data['mpi'])
mpi_dict = pd.Series(mpi_data.transformed_mpi.values, index=mpi_data.region).to_dict()

# Map transformed MPI scores to your labels
train_mpi_scores = np.array([mpi_dict[region] for region in train_labels])
test_mpi_scores = np.array([mpi_dict[region] for region in test_labels])

# Convert images to numpy arrays and transpose axes
train_images_np = np.array(train_images).transpose(0, 2, 3, 1)[:, :, :, :3]  # Keep only RGB channels
test_images_np = np.array(test_images).transpose(0, 2, 3, 1)[:, :, :, :3]

print(f"\nShape of training images: {train_images_np.shape}")
print(f"Shape of test images: {test_images_np.shape}")

# Compute mean and std for normalization (using training data only)
print("\nComputing mean and std for normalization from training data...")
all_pixels = np.concatenate([img.reshape(-1, 3) for img in train_images_np], axis=0)
mean = all_pixels.mean(axis=0) / 255.0
std = all_pixels.std(axis=0) / 255.0
print(f"Mean: {mean}, Std: {std}")

# Define data augmentation with normalization
data_augmentation = transforms.Compose([
    transforms.ToPILImage(),
    transforms.RandomRotation(90),
    transforms.RandomHorizontalFlip(),
    transforms.RandomVerticalFlip(),
    transforms.RandomResizedCrop(224, scale=(0.8, 1.0)),
    transforms.ColorJitter(brightness=(0.5, 1.5), contrast=(0.5, 1.5)),
    transforms.ToTensor(),
    transforms.Normalize(mean=mean, std=std)
])

# Define test transforms
test_transforms = transforms.Compose([
    transforms.ToPILImage(),
    transforms.Resize(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=mean, std=std)
])

# Apply augmentation to training images
print("\nApplying data augmentation to training images...")
augmented_images = []
augmented_mpi_scores = []
for idx, (img, label) in enumerate(zip(train_images_np, train_mpi_scores)):
    img = np.uint8(img)  # Ensure image is in uint8 format
    for _ in range(48):  # Generate augmentations per image
        aug_img = data_augmentation(img)
        augmented_images.append(aug_img.numpy())
        augmented_mpi_scores.append(label)
    if idx % 10 == 0:
        print(f"Augmented {idx+1}/{len(train_images_np)} images")

print(f"\nTotal augmented images: {len(augmented_images)}")
print(f"Total augmented MPI scores: {len(augmented_mpi_scores)}")

# Convert augmented images to numpy array
augmented_images_np = np.array(augmented_images)

# Split the augmented training data into training and validation sets
print("\nSplitting augmented data into training and validation sets...")
train_aug_images_np, val_aug_images_np, train_aug_mpi_scores, val_aug_mpi_scores = train_test_split(
    augmented_images_np, augmented_mpi_scores, test_size=0.2, random_state=42)

# Prepare tensors for PyTorch
train_aug_images_tensor = torch.tensor(train_aug_images_np, dtype=torch.float32)
train_aug_mpi_scores_tensor = torch.tensor(train_aug_mpi_scores, dtype=torch.float32).unsqueeze(1)

val_aug_images_tensor = torch.tensor(val_aug_images_np, dtype=torch.float32)
val_aug_mpi_scores_tensor = torch.tensor(val_aug_mpi_scores, dtype=torch.float32).unsqueeze(1)

# Process test images
print("\nProcessing test images...")
test_images_tensor = []
for img in test_images_np:
    img = test_transforms(np.uint8(img))
    test_images_tensor.append(img)

test_images_tensor = torch.stack(test_images_tensor)
test_mpi_scores_tensor = torch.tensor(test_mpi_scores, dtype=torch.float32).unsqueeze(1)

print(f"\nShape of training images tensor: {train_aug_images_tensor.shape}")
print(f"Shape of validation images tensor: {val_aug_images_tensor.shape}")
print(f"Shape of test images tensor: {test_images_tensor.shape}")

# Use GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"\nUsing device: {device}")

train_aug_images_tensor = train_aug_images_tensor.to(device)
train_aug_mpi_scores_tensor = train_aug_mpi_scores_tensor.to(device)
val_aug_images_tensor = val_aug_images_tensor.to(device)
val_aug_mpi_scores_tensor = val_aug_mpi_scores_tensor.to(device)
test_images_tensor = test_images_tensor.to(device)
test_mpi_scores_tensor = test_mpi_scores_tensor.to(device)

# Create datasets and dataloaders
train_dataset = TensorDataset(train_aug_images_tensor, train_aug_mpi_scores_tensor)
val_dataset = TensorDataset(val_aug_images_tensor, val_aug_mpi_scores_tensor)
test_dataset = TensorDataset(test_images_tensor, test_mpi_scores_tensor)

batch_size = 32
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

# Define the model
print("\nInitializing the model...")
# Load ResNet50 pretrained on ImageNet
model = models.resnet50(weights=ResNet50_Weights.IMAGENET1K_V2)

# Modify the model for regression
num_features = model.fc.in_features
model.fc = nn.Linear(num_features, 1)

model.to(device)

# Unfreeze layers from layer3 onwards
for name, param in model.named_parameters():
    if 'layer4' in name or 'layer3' in name or 'fc' in name:
        param.requires_grad = True
    else:
        param.requires_grad = False

# Loss function and optimizer
criterion = nn.SmoothL1Loss()  # Huber Loss
optimizer = optim.Adam(filter(lambda p: p.requires_grad, model.parameters()), lr=1e-4, weight_decay=1e-5)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=3)

# Training loop
num_epochs = 50
train_losses = []
val_losses = []
best_val_loss = float('inf')
patience = 7
patience_counter = 0

print("\nStarting training loop...")
for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    for inputs, labels in tqdm(train_loader, desc=f'Epoch {epoch+1}/{num_epochs} - Training'):
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
    train_loss = running_loss / len(train_loader)
    train_losses.append(train_loss)

    # Validation
    model.eval()
    running_val_loss = 0.0
    with torch.no_grad():
        for inputs, labels in val_loader:
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            running_val_loss += loss.item()
    val_loss = running_val_loss / len(val_loader)
    val_losses.append(val_loss)

    print(f'Epoch [{epoch+1}/{num_epochs}] - Train Loss: {train_loss:.6f}, Val Loss: {val_loss:.6f}')
    scheduler.step(val_loss)

    # Early stopping
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        patience_counter = 0
        # Save the model
        torch.save(model.state_dict(), 'best_model.pth')
    else:
        patience_counter += 1
        if patience_counter >= patience:
            print(f'Early stopping on epoch {epoch+1}')
            break

# Load the best model
model.load_state_dict(torch.load('best_model.pth'))

# Plot training and validation losses
plt.figure(figsize=(10, 5))
plt.plot(range(1, len(train_losses)+1), train_losses, label='Training Loss')
plt.plot(range(1, len(val_losses)+1), val_losses, label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.show()

# Evaluate on test set
print("\nEvaluating on test set...")
model.eval()
with torch.no_grad():
    predicted_mpi_scores_tensor = model(test_images_tensor)
    predicted_mpi_scores = predicted_mpi_scores_tensor.view(-1).cpu().numpy()

# De-transform predictions (square the outputs to reverse square root transformation)
predicted_mpi_scores = predicted_mpi_scores ** 2

# Aggregate predictions by region
print("\nAggregating predictions by region...")
unique_test_regions, test_region_indices = np.unique(test_labels, return_inverse=True)
agg_predicted_scores = np.array([predicted_mpi_scores[test_region_indices == i].mean() for i in range(len(unique_test_regions))])

# Actual MPI scores for test regions
actual_mpi_scores = np.array([mpi_data.loc[mpi_data['region'] == region, 'mpi'].values[0] for region in unique_test_regions])

# Compute metrics
mse = mean_squared_error(actual_mpi_scores, agg_predicted_scores)
rmse = np.sqrt(mse)
mae = np.mean(np.abs(agg_predicted_scores - actual_mpi_scores))
r2 = r2_score(actual_mpi_scores, agg_predicted_scores)

print(f"\nMetrics:")
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"R-squared (R²): {r2}")

# Plot actual vs predicted MPI scores
plt.figure(figsize=(10, 5))
plt.scatter(actual_mpi_scores, agg_predicted_scores)
plt.xlabel('Actual MPI Scores')
plt.ylabel('Predicted MPI Scores')
plt.title('Actual vs Predicted MPI Scores')
plt.plot([actual_mpi_scores.min(), actual_mpi_scores.max()], [actual_mpi_scores.min(), actual_mpi_scores.max()], 'r--')
plt.show()

# Save actual vs predicted MPI scores to a CSV file
scatter_data = pd.DataFrame({
    'Region': unique_test_regions,
    'Actual_MPI_Score': actual_mpi_scores,
    'Predicted_MPI_Score': agg_predicted_scores
})
scatter_data.to_csv('actual_vs_predicted_mpi_scores.csv', index=False)
print("\nScatter plot data saved to 'actual_vs_predicted_mpi_scores.csv'")

# Plot MPI scores on the map
print("\nPlotting MPI scores on the map...")
regions_shapes = admin_shapes[['ADM1_EN', 'geometry']].rename(columns={'ADM1_EN': 'Region'})
regions_shapes = regions_shapes.merge(mpi_data[['region', 'mpi']], left_on='Region', right_on='region', how='left')
predicted_df = pd.DataFrame({
    'Region': unique_test_regions,
    'Predicted_MPI': agg_predicted_scores
})
regions_shapes = regions_shapes.merge(predicted_df, on='Region', how='left')

vmin = min(regions_shapes['mpi'].min(), regions_shapes['Predicted_MPI'].min())
vmax = max(regions_shapes['mpi'].max(), regions_shapes['Predicted_MPI'].max())

fig, axes = plt.subplots(1, 2, figsize=(24, 12))

# Actual MPI scores
regions_shapes.plot(ax=axes[0], column='mpi', cmap='OrRd', vmin=vmin, vmax=vmax, edgecolor='black')
axes[0].set_title('Actual MPI Scores')
axes[0].axis('off')

# Predicted MPI scores
regions_shapes.plot(ax=axes[1], column='Predicted_MPI', cmap='OrRd', vmin=vmin, vmax=vmax, edgecolor='black')
axes[1].set_title('Predicted MPI Scores')
axes[1].axis('off')

# Colorbar
fig.colorbar(plt.cm.ScalarMappable(cmap='OrRd', norm=plt.Normalize(vmin=vmin, vmax=vmax)),
             ax=axes.ravel().tolist(), orientation='horizontal', fraction=0.07, pad=0.07, label='MPI Score')

plt.show()
```

```python
# Sampling 10 new images per region and making predictions
print("\nSampling 10 new images per region and making predictions...")

# Initialize lists to store predictions and actual MPI scores
predicted_mpi_scores = []
actual_mpi_scores = []
regions_with_preds = []

# Use the same device as before
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Adjusted extract_images function (without labels)
def extract_images(coords, image_path, size=224):
    images = []
    with rasterio.open(image_path) as src:
        for coord in coords:
            row, col = src.index(coord[0], coord[1])
            window = Window(col - size // 2, row - size // 2, size, size)
            try:
                image = src.read(window=window)
                if image.shape[1] != size or image.shape[2] != size:
                    print(f"Skipping image at {coord} due to incorrect size {image.shape}")
                    continue
                images.append(image)
            except Exception as e:
                print(f"Error reading image at {coord}: {e}")
                continue
    return images

# Iterate over all regions
for region in regions:
    print(f"\nProcessing region: {region}")
    # Get the geometry for the region
    geometry = admin_shapes.loc[admin_shapes['ADM1_EN'] == region, 'geometry'].values[0]
    
    # Sample 10 new coordinates within the region
    new_coords = sample_unique_coordinates(geometry, 10)
    
    # Prepare coordinate list
    coord_list = [(point.x, point.y) for point in new_coords]
    
    # Extract images at those coordinates
    images = extract_images(coord_list, sat_image_path)
    
    if len(images) == 0:
        print(f"No valid images found for region {region}. Skipping.")
        continue
    
    # Process images with test transforms
    processed_images = []
    for img in images:
        img_np = img.transpose(1, 2, 0)[:, :, :3]  # Convert to H x W x C and keep RGB channels
        img_pil = test_transforms(np.uint8(img_np))
        processed_images.append(img_pil)
    
    # Stack images into a tensor
    images_tensor = torch.stack(processed_images).to(device)
    
    # Make predictions using the trained model
    model.eval()
    with torch.no_grad():
        preds = model(images_tensor)
        preds = preds.view(-1).cpu().numpy()
        # De-transform predictions
        preds = preds ** 2  # Reverse the square root transformation
    
    # Aggregate predictions per region
    agg_pred = preds.mean()
    predicted_mpi_scores.append(agg_pred)
    
    # Get actual MPI score
    if region in mpi_data['region'].values:
        actual_score = mpi_data.loc[mpi_data['region'] == region, 'mpi'].values[0]
        actual_mpi_scores.append(actual_score)
        regions_with_preds.append(region)
    else:
        print(f"No actual MPI score for region {region}. Skipping.")

# Compute metrics
mse = mean_squared_error(actual_mpi_scores, predicted_mpi_scores)
rmse = np.sqrt(mse)
mae = np.mean(np.abs(np.array(predicted_mpi_scores) - np.array(actual_mpi_scores)))
r2 = r2_score(actual_mpi_scores, predicted_mpi_scores)

print(f"\nMetrics on new samples:")
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"R-squared (R²): {r2}")

# Plot actual vs predicted MPI scores
plt.figure(figsize=(10, 5))
plt.scatter(actual_mpi_scores, predicted_mpi_scores)
plt.xlabel('Actual MPI Scores')
plt.ylabel('Predicted MPI Scores')
plt.title('Actual vs Predicted MPI Scores on New Samples')
plt.plot([min(actual_mpi_scores), max(actual_mpi_scores)], [min(actual_mpi_scores), max(actual_mpi_scores)], 'r--')
plt.show()

# Save actual vs predicted data to CSV
scatter_data = pd.DataFrame({
    'Region': regions_with_preds,
    'Actual_MPI_Score': actual_mpi_scores,
    'Predicted_MPI_Score': predicted_mpi_scores
})
scatter_data.to_csv('actual_vs_predicted.csv', index=False)
print("\nScatter plot data saved to 'actual_vs_predicted.csv'")

# Plot MPI scores on the map
print("\nPlotting MPI scores on the map for new predictions...")
regions_shapes = admin_shapes[['ADM1_EN', 'geometry']].rename(columns={'ADM1_EN': 'Region'})
regions_shapes = regions_shapes.merge(mpi_data[['region', 'mpi']], left_on='Region', right_on='region', how='left')
predicted_df = pd.DataFrame({
    'Region': regions_with_preds,
    'Predicted_MPI': predicted_mpi_scores
})
regions_shapes = regions_shapes.merge(predicted_df, on='Region', how='left')

vmin = min(regions_shapes['mpi'].min(), regions_shapes['Predicted_MPI'].min())
vmax = max(regions_shapes['mpi'].max(), regions_shapes['Predicted_MPI'].max())

fig, axes = plt.subplots(1, 2, figsize=(24, 12))

# Actual MPI scores
regions_shapes.plot(ax=axes[0], column='mpi', cmap='OrRd', vmin=vmin, vmax=vmax, edgecolor='black')
axes[0].set_title('Actual MPI Scores')
axes[0].axis('off')

# Predicted MPI scores
regions_shapes.plot(ax=axes[1], column='Predicted_MPI', cmap='OrRd', vmin=vmin, vmax=vmax, edgecolor='black')
axes[1].set_title('Predicted MPI Scores on New Samples')
axes[1].axis('off')

# Colorbar
fig.colorbar(plt.cm.ScalarMappable(cmap='OrRd', norm=plt.Normalize(vmin=vmin, vmax=vmax)),
             ax=axes.ravel().tolist(), orientation='horizontal', fraction=0.07, pad=0.07, label='MPI Score')

plt.show()
```
